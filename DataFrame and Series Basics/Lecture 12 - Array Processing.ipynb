```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
points = np.arange(-5,5,0.01)
```

```python
#Points by points grid
dx,dy = np.meshgrid(points,points)
```

```python
#Matrix where all the rows go from -5 to +5
dx
```

```python
#Same thing except as long as through the columns
dy
```

```python
z = (np.sin(dx) + np.sin(dy))

z
```

```python
#Image show (good way to show this data, basically a matrix, a sin wave along x and y's)
plt.imshow(z)
```

```python
plt.imshow(z)
plt.colorbar()

plt.title('Plot for sin(x) + sin(y)')
```

```python
# Numpy where

A = np.array([1,2,3,4])

B = np.array([100,200,300,400])
```

```python
#Boolean Array
condition = np.array([True,True,False,False])
```

```python
#List comprehension
answer = [(A_val if cond else B_val)for A_val,B_val,cond in zip(A,B,condition)]

#Chooses the A_val if cond is true, else choose the B_val
# maps A_val,B_val,cond to A,B and condition
```

```python
answer
```

```python
#Short version of the list comprehension 
answer2 = np.where(condition,A,B)

answer2
```
