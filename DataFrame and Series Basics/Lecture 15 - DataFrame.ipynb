```python
import numpy as np

import pandas as pd

from pandas import Series,DataFrame
```

```python
#Pythons built in method to grab a webpage
import webbrowser

website = 'http://en.wikipedia.org/wiki/NFL_win-loss_records'
webbrowser.open(website)
```

```python
#Copied first 5 rows from website to create dataframe
nfl_frame = pd.read_clipboard()
#Pandas built in function to read everything from clipboard and set it to a dataframe
```

```python
nfl_frame
```

```python
nfl_frame.columns
```

```python
nfl_frame.Rank
```

```python
#How to grab a column (one word)
nfl_frame.Team
```

```python
#More than one word
nfl_frame['First NFL season']
```

```python
#Grab multiple columns
DataFrame(nfl_frame,columns=['Team','First NFL season','GP'])
```

```python
#What happens when you call a column that doesn't exist?
DataFrame(nfl_frame,columns=['Team','First NFL season','GP','Stadium'])
```

Pandas is great because it doesn't return an error but returns the column with Null values
The dataframe still works
