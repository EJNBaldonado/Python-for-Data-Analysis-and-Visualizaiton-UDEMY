Basic Statistics Notes
Bivariate relationships (2-variable relationships)
COVARIANCE
Real Ex: DJI vs S&P 500 Graph of the overall performance of stock market
Linear Pattern - Positive Linear Relationship 
when one variable moves in a direction the other tends to move in the same direction THIS IS CALLED COVARIANCE!
A descriptive measure of the linear association between two variables
+(both goes in same direction) -(go in opposite directions)

Correlation talks about the STRENGTH!
Variables with no real linear relationship will probably have a 0 or near 0 covariance
COVARIANCE FORMULAS

IGNORE THE SIZE OF THE NUMBER - JUST LOOK AT THE SIGN! (No measure of strength)

CORRELATION
Covariance has no upper or lower bound - size dependent on scale of variables (not standardized)
CORRELATION is ALWAYS between -1 and +1 and scale is independent of scale of variables (standardized think z-score)
Correlation is ONLY applicable to LINEAR relationships. Many other types of relationships that can exist between two variables (U-shaped etc.)

CORRELATION IS NOT CAUSATION
r is called the (Pearson) correlation coefficient
r = Cov(x,y) / Std.Dev(x) * Std.Dev(y) 
CLOSER TO 1/-1 IS STRONGER, CLOSER TO 0 IS WEAKER RELATIONSHIP

Covariance provides the direction - Correlation provides direction and strength

```python
import pandas_datareader as pdweb
import datetime
```

```python
prices = pdweb.get_data_yahoo(['CVX','XOM','BP'],start=datetime.datetime(2010,1,1),
                             end=datetime.datetime(2013,1,1))['Adj Close']
```

```python
prices.head()
```

```python
volume = pdweb.get_data_yahoo(['CVX','XOM','BP'],start=datetime.datetime(2010,1,1),
                             end=datetime.datetime(2013,1,1))['Volume']
```

```python
volume.head()
```

```python
#Looking for the returns on your stock, can use percent change
rets = prices.pct_change()
```

```python
#Look at the correlations of your returns
corr = rets.corr
```

```python
%matplotlib inline
prices.plot()
```
