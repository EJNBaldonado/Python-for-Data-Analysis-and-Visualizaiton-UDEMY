```python
dataset1 = randn(100)
```

```python
# Create a histogram, by default bins into 10 equally spaced intervals
plt.hist(dataset1)
```

```python
dataset2 = randn(80)
plt.hist(dataset2, color = 'indianred')
```

```python
# Plotting both datasets in the same chart
# density = True instead of normed since normed is no longer supported
# alpha is transparancy
plt.hist(dataset1, density=True, color='indianred', alpha=0.5, bins=20)
plt.hist(dataset2, density=True, alpha=0.5, bins=20)
```
