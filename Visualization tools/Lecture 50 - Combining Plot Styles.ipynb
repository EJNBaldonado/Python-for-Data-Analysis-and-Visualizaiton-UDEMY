```python
#The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd
#Stats
from scipy import stats
#Plotting libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
#to see plots in notebook
%matplotlib inline
```

```python
dataset = randn(100)
```

```python
# Use distplot to combine plot
sns.distplot(dataset,bins=25)
```

```python
#Since distplot is being deprecated, lets see other functions
sns.displot(dataset,bins=25)
```

```python
#Seems like this is the best fit!
sns.histplot(dataset,bins=25,kde=True)
```

```python
sns.distplot(dataset,bins=25,rug=True,hist=False)
```

```python
#Alternate using displot
sns.displot(dataset,kind='kde',rug=True)
```

```python
sns.distplot(dataset,bins=25,
            kde_kws={'color':'indianred','label':'KDE PLOT'},
            hist_kws={'color':'blue','label':'HIST'})
```

```python
sns.histplot(dataset,bins=25,kde=True,
            line_kws={'color':'indianred','label':'KDE PLOT'},
            color='blue', label='HIST')
```
