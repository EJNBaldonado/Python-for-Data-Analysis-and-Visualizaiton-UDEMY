Probability Density Function (PDF) vs Boxplot


```python
data1 = randn(100)
data2 = randn(100)
```

```python
# Making boxplots using seaborns
sns.boxplot(data=[data1,data2])
```

```python
# Include outliers in whiskers
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
# Setting boxplot horizontally, vert = False now orient (for orientation either 'h' or 'v')
sns.boxplot(data=data1,orient='h')
```

```python
# Violin / Viola plot (combines simplicity of boxplot with the info from a KDE plot)
# Normal Dist
data1 = stats.norm(0,5).rvs(100)
# Two Gamma dist. concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1*stats.gamma(5).rvs(50)])

# Boxplot both data1 and data2
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
# With the kde plot you can see the difference between the two
# 1st clearly being a normal distribution, while the 2nd one showing the 2 gammas one neg one pos
sns.violinplot(data=[data1,data2])
```

```python
# Since the violin plot includes a kde plot
# you can adjust the bandwidth of the kernel in the violin plot
sns.violinplot(data=data2, bw=0.01)
```

```python
# Similar to rug plot you can include the individual points
sns.violinplot(data=data1,inner='stick')
```
